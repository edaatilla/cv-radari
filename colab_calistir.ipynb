```python
# 5) Uygulamayı başlat ve genel linki üret
import subprocess, time, re

server = subprocess.Popen(["python", "app.py"])
time.sleep(8)

tunnel = subprocess.Popen(
    ["./cloudflared", "tunnel", "--url", "http://localhost:7860"],
    stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True,
)

print("Link hazirlaniyor, birkac saniye bekle...\n")
for line in tunnel.stdout:
    match = re.search(r"https://[a-zA-Z0-9\-]+\.trycloudflare\.com", line)
    if match:
        print("=== UYGULAMA LINKIN HAZIR ===")
        print(match.group(0))
        print("==============================")
        print("\nBu linki istediğin kişiyle paylaşabilirsin (bu Colab defteri açık kaldığı sürece çalışır).")
        break
```
